#แผนที่แสดงพรรคที่ได้รับคะแนนโหวตมากที่สุดในแต่ละจังหวัด

```python
find_ma = result_constituencies_PartyList.sort_values(by='party_list_vote',ascending=False)
find_ma1 = find_ma.merge(info_constituency,how='left', on='cons_id')
find_ma2 = find_ma1.merge(info_province,how='left', on='prov_id')
find_ma3 = find_ma2.merge(info_party_overview,how='left', left_on='party_id',right_on='id')
find_ma4 = find_ma3.merge(Candidate_PartyList,how='left', on='party_no')
```

```python
find_mashow = find_ma4[['province','name_x','name_y','party_list_vote']]
find_mashow['party_list_vote'] = find_ma4[['party_list_vote']]
find_mashow
```

```python
use_V0=find_mashow[['name_x','name_y','party_list_vote']]
prov_vote=find_mashow.groupby('province')[['party_list_vote']].max().reset_index()
use_V1 = prov_vote.merge(use_V0,how='left', on='party_list_vote')
bool_use = use_V1.duplicated(subset='province')
use_V1 = use_V1[~bool_use]
# displaying data
use_V1.info()
use_V1
```
